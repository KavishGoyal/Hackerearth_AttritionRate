```python
target = df['Attrition_rate']
df = df.drop('Attrition_rate', 1)
df.head()
```

```python
categorical_feat = [feat for feat in df.columns if df[feat].dtype == 'object' ]
```

```python
numeric_feat = [feat for feat in df.columns if df[feat].dtype != 'object' ]
```

```python
numeric_feat, categorical_feat
```

```python
! pip3 install category_encoders
```

```python
#from sklearn.preprocessing import LabelEncoder,  CategoricalEncoders
#from sklearn.preprocessingimport CategoricalEncoder
#le = LabelEncoder()
import category_encoders as ce

from category_encoders.target_encoder import TargetEncoder
te = TargetEncoder()
#ce = CategoricalEncoder()
cat_features = te.fit_transform(df[categorical_feat], target)
```

```python
cat_features.head(), df[numeric_feat].head()
```

```python
df_new = pd.concat([df[numeric_feat], cat_features])
```

```python
#transform the numeric features using log(x + 1)
from scipy.stats import skew
skewed = df[numeric_feat].apply(lambda x: skew(x.dropna().astype(float)))
skewed = skewed[skewed > 0.75]
skewed = skewed.index
df[skewed] = np.log1p(df[skewed])
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
num_pipeline = Pipeline([
#('imputer', SimpleImputer(strategy="median")),

('std_scaler', StandardScaler()),
])
df_num_tr = num_pipeline.fit_transform(df[numeric_feat])
```

```python
df_num_tr.shape, cat_features.shape
```

```python
cat_features.drop('Employee_ID', 1)
```

```python
cat_feats = np.array(cat_features.values)
```

```python
df_new = np.concatenate([df_num_tr, cat_feats], axis = 1)
```

```python
df_new.shape
```

```python
from sklearn.model_selection import train_test_split
# split train data
x_train, x_test, y_train, y_test = train_test_split(
    df_new, target,  
    test_size=0.2, 
    random_state=42
)
```

```python
len(x_train), len(y_train), len(x_test), len(y_test)
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_rgr = RandomForestRegressor(random_state=42)
forest_rgr.fit(x_train, y_train)
frst_prd = forest_rgr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, frst_prd)
print(mse)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, frst_prd)
```

```python
# traing the data by svm classifier
from sklearn.svm import SVR
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import reciprocal, uniform

param_distributions = {"gamma": reciprocal(0.001, 0.1), "C": uniform(1, 10)} 
rnd_search_cv = RandomizedSearchCV(SVR(), param_distributions, n_iter=10, verbose=2, cv=5, random_state=42)
rnd_search_cv.fit(x_train, y_train)
```

```python
rnd_search_cv.best_estimator_
```

```python

y_pred = rnd_search_cv.best_estimator_.predict(x_test)
```

```python
y_pred[:10]
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print(mse)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
# FIRST we flatten the data and then used 2 hidden layers each with 3 neurons and relu activation function.
# Then the final output layer with 2 neuron(as we have binary classes  our output) with softmax activation function.
import tensorflow as tf
model = tf.keras.models.Sequential([tf.keras.layers.Flatten(), 
                                    tf.keras.layers.Dense(3, activation=tf.nn.relu),  
                                    tf.keras.layers.Dense(3, activation=tf.nn.relu),
                                    tf.keras.layers.Dense(1, activation=tf.nn.softmax)])
```

```python
# Now compile the model with 'adam' optimizer and categorical loss function.
model.compile(optimizer = 'Adam',
              loss = 'categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Now train the model using spited dataset and get the loss and accuracy as score.
batch_size = 80    
history = model.fit(x_train, y_train, batch_size=batch_size, epochs=10, validation_data = (x_test,y_test), verbose=2)
```

```python
from sklearn.ensemble import AdaBoostRegressor
abr = AdaBoostRegressor(n_estimators = 10)
abr.fit(x_train, y_train)
pred = abr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, pred)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, pred)
print(mse)
```
